```python
import seaborn as sns
import numpy as np
from scipy.stats import norm
import statistics
import pandas as pd
import matplotlib.pyplot as plt
import math
from matplotlib.patches import Polygon
import seaborn as sb


# crear distribucion desde 4.8 a 5.35 en incrementos de 0.01
x_axis = np.arange(4.8, 5.35, 0.01)
  
# Media y Desviacion Estandar provistas
mean = 5.07
sd = .07
  
# Curva de distribucion    
fig, ax = plt.subplots()
ax.plot(x_axis, norm.pdf(x_axis, mean, sd) , color='#CD5430', alpha=1.00)
ax.fill_between(x_axis,norm.pdf(x_axis, mean, sd), 0, color='#CD5430', alpha=.1)
plt.title('Distribucion')
plt.axvline(x = mean, color = 'r', label = '')
ax.set_yticks([])
ax.set_xticks([mean])
```

```python
import math
def normpdf(x, mean, sd):
    var = float(sd)**2
    denom = (2*math.pi*var)**.5
    num = math.exp(-(float(x)-float(mean))**2/(2*var))
    return num/denom
normpdf(5.05,5.07,0.07)  
```

```python
'''
Calcular la integral definida, desde el limite inferior 5.07,
hasta el limite superior 5.15, restar las probabilidades.
'''

#limites del poligono para el area bajo la curva
a, b = 5.07, 5.15
x = x_axis
y = norm.pdf(x_axis, mean, sd)

#probabilidad
cdf_upper_limit = norm(loc = mean , scale = sd).cdf(b)
cdf_lower_limit = norm(loc = mean , scale = sd).cdf(a)
 
prob = cdf_upper_limit - cdf_lower_limit
print(f'La probabilidad de que las varillas entre\n\
{a} y {b} del inventario se ajusten es de: {round(prob*100,2)} %')

# definir f(x)
def func(x):
    return norm.pdf(x, mean, sd)

#grafico
fig, ax = plt.subplots()
ax.plot(x, y, '#CD5430', linewidth=2)
ax.set_ylim(bottom=0)

# sombreado de area
ix = np.linspace(a, b)
iy = func(ix)
verts = [(a, 0), *zip(ix, iy), (b, 0)]
poly = Polygon(verts, facecolor='#F9C1B3', edgecolor='#CD5430')
ax.add_patch(poly)

fig.text(0.9, 0.05, '$x$')
fig.text(0.1, 0.9, '$y$')
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
ax.xaxis.set_ticks_position('bottom')
plt.title('Actividad 2, limite superior')
ax.set_yticks([])
ax.set_xticks([a, b])
plt.show()
```
